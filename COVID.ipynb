```python
gdf = gpd.GeoDataFrame(
    covid, geometry=gpd.points_from_xy(covid.Long, covid.Lat))
```

```python
gdfheaderlist = gdf.columns.tolist()
len(gdfheaderlist)-1
```

```python
DateN=gdf.columns.tolist()
for i in range(70,len(gdfheaderlist)-1): 
    fig, ax = plt.subplots(figsize=(12, 8))
    #ax.set(title='Corona virus cases')
    

    worldBound.plot(figsize=(10, 5), color='lightgrey',
               ax=ax)
    # remove the axis
    ax.axis('off')
    date_object = datetime.strptime(DateN[i], '%m/%d/%y')
    date_ready = date_object.strftime("%b %d")
    
    # title
    ax.annotate('COVID-19 cases',
               xy=(0.09, .87), xycoords='figure fraction',
               horizontalalignment='left', verticalalignment='top',
               fontsize=25, fontweight='bold', color='#555555')
    
    #date
    ax.annotate(date_ready,
               xy=(0.09, .63), xycoords='figure fraction',
               horizontalalignment='left', verticalalignment='top',
               fontsize=20, color='#555555')
    
    #known cases
    ax.annotate(f'{gdf.sum(axis=0)[i-1]:,}''\nknown cases',
               xy=(0.09, .565), xycoords='figure fraction',
               horizontalalignment='left', verticalalignment='top',
               fontsize=20, color='#555555')
    
    #source/byline
    ax.annotate('Source: Johns Hopkins University\nRandy Leonard/CQ Roll Call',
               xy=(0.09, .20), xycoords='figure fraction',
               horizontalalignment='left', verticalalignment='top',
               fontsize=12, color='#555555')

    # Add outbreak country/state locations
    size=gdf.iloc[:, [i]]
    gdf.plot(ax=ax, 
                color='coral', alpha=.6, 
                markersize=(size**.5)*3)
    
    #key
    ax.scatter([-90], [42], c='lightgray', alpha=1, zorder=-1, s=1**.5*3,
            label='1')
    ax.scatter([-90], [42], c='lightgray', alpha=.6, zorder=-1, s=500**.5*3,
            label='500')
    ax.scatter([-90], [42], c='lightgray', alpha=.6, zorder=-1, s=10000**.5*3,
            label='10,000')
    ax.scatter([-90], [42], c='lightgray', alpha=.6, zorder=-1, s=50000**.5*3,
            label='50,000')

    ax.legend( frameon=False,
        labelspacing=0.7, loc='lower left', fontsize=14, 
        bbox_to_anchor=(0.05,0.05), title_fontsize=20);
    #addlogo
    logo = mpimg.imread('cqrclogo.png')
    imagebox = OffsetImage(logo, zoom=.4)

    ab = AnnotationBbox(imagebox, (150, -62),frameon=False)

    ax.add_artist(ab)

    
    fig.savefig('testmap'+"{:0>2d}".format(i)+'.png', dpi=300, bbox_inches = 'tight',
    pad_inches = 0)

#with open('file_{0}.dat'.format(i),'w') as f:
       #f.write(str(func(i)))
    



    


# Setup x y axes with labels and add graticules
```
